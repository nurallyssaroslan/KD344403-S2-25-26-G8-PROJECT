```python
from sklearn.model_selection import KFold  # Import KFold for cross-validation splitting
import matplotlib.pyplot as plt  # Import library for creating charts
from tqdm.notebook import tqdm  # Import tqdm progress bar for Google Colab / Jupyter Notebook

# Create an empty dictionary to store the R² scores for each fold of each model
fold_results = {}

# Define 7-fold cross-validation
# shuffle=True randomizes the data before splitting
# random_state=42 ensures the same folds are produced every time the code is run
kf = KFold(n_splits=7, shuffle=True, random_state=42)

# Loop through each model in the models dictionary
# tqdm shows the overall progress for all models
for name, model in tqdm(models.items(), total=len(models), desc="Running 7-Fold CV"):

    # Create an empty list to store the R² scores for the current model
    scores = []

    # Loop through each fold generated by KFold
    # tqdm shows the progress for the current model across 7 folds
    for train_idx, val_idx in tqdm(
        kf.split(X_train),
        total=7,
        desc=f"{name}",
        leave=False
    ):
        # Split the training data into training fold and validation fold
        X_tr, X_val = X_train.iloc[train_idx], X_train.iloc[val_idx]
        y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]

        # Train the model using the training fold
        model.fit(X_tr, y_tr)

        # Evaluate the model on the validation fold using R² score
        score = model.score(X_val, y_val)

        # Store the R² score for this fold
        scores.append(score)

    # Store all 7 fold scores for the current model
    fold_results[name] = scores

# Set the figure size for the line chart
plt.figure(figsize=(12, 6))

# Plot the R² score for each fold for every model
for name, scores in fold_results.items():
    plt.plot(range(1, 8), scores, marker='o', label=name)

# Add chart title
plt.title('R² Score Per Fold (7-Fold Cross-Validation)')

# Label the x-axis
plt.xlabel('Fold')

# Label the y-axis
plt.ylabel('R² Score')

# Set x-axis values from Fold 1 to Fold 7
plt.xticks(range(1, 8))

# Display the legend outside the chart to avoid covering the lines
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')

# Add light grid lines to make score differences easier to read
plt.grid(alpha=0.3)

# Adjust layout so labels and legend are not cut off
plt.tight_layout()

# Display the chart
plt.show()
```
