```python
# Import regression models from sklearn
from sklearn.linear_model import LinearRegression, Ridge, Lasso

# Import ensemble model
from sklearn.ensemble import RandomForestRegressor

# Import Support Vector Regression model
from sklearn.svm import SVR

# Import boosting models
from xgboost import XGBRegressor
from lightgbm import LGBMRegressor
from catboost import CatBoostRegressor

# Dictionary containing all regression models
models = {

    # Basic Linear Regression model
    "LinearRegression": LinearRegression(),

    # Ridge Regression (L2 Regularization)
    "Ridge": Ridge(),

    # Lasso Regression (L1 Regularization)
    "Lasso": Lasso(),

    # Random Forest Regressor
    # random_state=42 ensures consistent results every run
    "RandomForestRegressor": RandomForestRegressor(random_state=42),

    # Support Vector Regressor
    "SVR": SVR(),

    # XGBoost Regressor
    # random_state=42 ensures reproducibility
    # verbosity=0 hides unnecessary training messages
    "XGBoost": XGBRegressor(random_state=42, verbosity=0),

    # LightGBM Regressor
    # random_state=42 ensures consistent results
    "LightGBM": LGBMRegressor(random_state=42),

    # CatBoost Regressor
    # random_state=42 ensures reproducibility
    # verbose=0 hides training output
    "CatBoost": CatBoostRegressor(random_state=42, verbose=0)
}
```
