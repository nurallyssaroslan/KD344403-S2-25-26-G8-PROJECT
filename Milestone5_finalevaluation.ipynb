```python
# Import necessary libraries
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from catboost import CatBoostRegressor

# Final tuned CatBoost model
final_model = CatBoostRegressor(
    depth=10,           # tree depth
    iterations=600,     # number of boosting iterations
    l2_leaf_reg=5,      # L2 regularization term
    learning_rate=0.07, # step size
    rsm=0.8,            # random subspace method: % of features per tree
    subsample=0.6,      # % of data sampled per tree
    verbose=0,          # suppress CatBoost output
    random_state=42,    # ensures reproducible results
    allow_writing_files=False
)

# Train final model on 80% training set
final_model.fit(X_train, y_train)

# Predict on the 20% test set
y_pred_final = final_model.predict(X_test)

# Final test metrics
test_mse = mean_squared_error(y_test, y_pred_final)
test_rmse = np.sqrt(test_mse)
test_r2 = r2_score(y_test, y_pred_final)

print("Final Evaluation on 20% Test Set")
print("Test R²:", round(test_r2, 4))
print("Test MSE:", round(test_mse, 4))
print("Test RMSE:", round(test_rmse, 4))

# Results table
results_table = pd.DataFrame({
    "Metric": ["Test R²", "Test MSE", "Test RMSE"],
    "Result": [test_r2, test_mse, test_rmse]
})

results_table["Result"] = results_table["Result"].round(4)
display(results_table)
```

```python
# Create a new figure for prediction performance plot
plt.figure(figsize=(10, 6))

# Plot actual values against predicted values
# Each point represents one test sample
plt.scatter(y_test, y_pred_final, alpha=0.6, label="Predictions")

# Draw perfect prediction line
# If the model predicted perfectly, all points would lie on this dashed line
plt.plot(
    [y_test.min(), y_test.max()],
    [y_test.min(), y_test.max()],
    linestyle="--",
    label="Perfect Prediction Line"
)

# Add title and axis labels
plt.title("Final Prediction Performance on Test Set")
plt.xlabel("Actual Log Salary")
plt.ylabel("Predicted Log Salary")

# Show legend and grid for easier interpretation
plt.legend()
plt.grid(alpha=0.3)

# Adjust layout to prevent label/title overlap
plt.tight_layout()

# Display the plot
plt.show()
```

```python
# Calculate residuals
# Residual = actual value - predicted value
residuals = y_test - y_pred_final

# Create a new figure for the residual plot
plt.figure(figsize=(10, 6))

# Plot predicted values against residuals
# Helps check whether the model errors are randomly distributed
plt.scatter(y_pred_final, residuals, alpha=0.6)

# Add a horizontal reference line at zero
# Points close to this line indicate smaller prediction errors
plt.axhline(0, linestyle="--")

# Add axis labels and title
plt.xlabel("Predicted Log Salary")
plt.ylabel("Residuals")
plt.title("Residual Plot on Test Set")

# Show grid for easier reading
plt.grid(alpha=0.3)

# Adjust layout to prevent overlap
plt.tight_layout()

# Display the residual plot
plt.show()
```
