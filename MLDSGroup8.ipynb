**7-FOLD CROSS VALIDATION**

```python
# Install tqdm library to show progress bars during model training
pip install tqdm
```

```python
# Import required libraries
from sklearn.model_selection import cross_val_score, KFold
from tqdm import tqdm
import numpy as np

# Create an empty dictionary to store cross-validation results for each model
cv_results = {}

# Set up 7-fold cross-validation
# shuffle=True randomizes the data before splitting
# random_state=42 makes the result reproducible
kf = KFold(n_splits=7, shuffle=True, random_state=42)

# Loop through each model in the models dictionary
for name, model in models.items():
    print(f"\nTraining {name}...")

    # Store R² scores from each fold
    scores = []

    # Run 7-fold cross-validation manually
    # tqdm shows the training progress and estimated remaining time
    for train_idx, val_idx in tqdm(kf.split(X_train), total=7):

        # Split the training data into training and validation sets for this fold
        X_tr, X_val = X_train.iloc[train_idx], X_train.iloc[val_idx]
        y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]

        # Train the model using the training fold
        model.fit(X_tr, y_tr)

        # Evaluate the model on the validation fold using R² score
        score = model.score(X_val, y_val)

        # Save the R² score for this fold
        scores.append(score)

    # Store the mean and standard deviation of the 7-fold CV R² scores
    cv_results[name] = {
        "CV_R2_Mean": np.mean(scores),
        "CV_R2_STD": np.std(scores)
    }

    # Print the average R² score across all folds
    print(f"{name} CV R2 Mean: {np.mean(scores):.4f}")

    # Print how much the R² score varies across folds
    print(f"{name} CV R2 Std: {np.std(scores):.4f}")
```

**MODEL COMPARISON**

```python
# Import evaluation metrics
from sklearn.metrics import mean_squared_error, r2_score

# Dictionary to store evaluation results for each model
test_results = {}

# Loop through each model in the models dictionary
for name, model in models.items():

    # Train the model using training data
    model.fit(X_train, y_train)

    # Predict target values using test data
    y_pred = model.predict(X_test)

    # Calculate Mean Squared Error (MSE)
    mse = mean_squared_error(y_test, y_pred)

    # Calculate R-squared (R2) score
    r2 = r2_score(y_test, y_pred)

    # Store evaluation metrics in dictionary
    test_results[name] = {
        "MSE": mse,
        "R2": r2
    }

# Display all model evaluation results
print(test_results)


```

```python
# Create an empty list to store final evaluation results
final_results = []

# Loop through all models
for name in models.keys():

    # Retrieve cross-validation mean R2 score
    cv_r2 = cv_results[name]["CV_R2_Mean"]

    # Retrieve cross-validation standard deviation
    cv_std = cv_results[name]["CV_R2_STD"]

    # Retrieve test set R2 score
    test_r2 = test_results[name]["R2"]

    # Retrieve Mean Squared Error (MSE)
    mse = test_results[name]["MSE"]

    # Calculate difference between CV R2 and Test R2
    gap = cv_r2 - test_r2

    # Calculate Root Mean Squared Error (RMSE)
    rmse = np.sqrt(mse)

    # Determine model performance status
    # Underfitting: both CV and Test R2 are very low
    if cv_r2 < 0.2 and test_r2 < 0.2:
        status = "Underfitting"

    # Overfitting: large gap between CV and Test performance
    elif gap > 0.05:
        status = "Overfitting"

    # Otherwise model is considered a good fit
    else:
        status = "Good Fit"

    # Store all evaluation results in dictionary format
    final_results.append({
        "Model": name,
        "CV_R2": round(cv_r2, 4),
        "CV_R2_STD": round(cv_std, 4),
        "Test_R2": round(test_r2, 4),
        "MSE": round(mse, 4),
        "RMSE": round(rmse, 4),
        "Gap(CV-Test)": round(gap, 4),
        "Status": status
    })

# Convert results list into pandas DataFrame
df_results = pd.DataFrame(final_results)

# Sort models based on highest Test R2 score
df_results = df_results.sort_values(by="Test_R2", ascending=False)

# Display final comparison table
print(df_results)
```

```python
from catboost import CatBoostRegressor  # Import CatBoost regression model

models = {  # Dictionary to store the model used in the project
    # CatBoost Regressor model
    # random_state=42 makes the result reproducible
    # verbose=0 hides detailed training output
    "CatBoost": CatBoostRegressor(random_state=42, verbose=0)}
```

```python

from sklearn.model_selection import KFold  # Import KFold to split training data into 7 folds
import matplotlib.pyplot as plt  # Import matplotlib to plot the cross-validation graph
from tqdm.notebook import tqdm   # for Google Colab / Jupyter Notebook  # Shows progress bar during CV process

fold_results = {}  # Dictionary to store R² scores for each model across all folds

kf = KFold(n_splits=7, shuffle=True, random_state=42)  # Create 7-fold CV with shuffled data and fixed random state

for name, model in tqdm(models.items(), total=len(models), desc="Running 7-Fold CV"):  # Loop through each model with progress bar
    scores = []  # List to store R² scores for the current model

    for train_idx, val_idx in tqdm(  # Loop through each fold split
        kf.split(X_train),  # Split X_train into training and validation indexes
        total=7,  # Total number of folds
        desc=f"{name}",  # Show current model name in progress bar
        leave=False  # Hide inner progress bar after completion
    ):
        X_tr, X_val = X_train.iloc[train_idx], X_train.iloc[val_idx]  # Split features into training and validation fold
        y_tr, y_val = y_train.iloc[train_idx], y_train.iloc[val_idx]  # Split target into training and validation fold

        model.fit(X_tr, y_tr)  # Train model using the training fold
        score = model.score(X_val, y_val)  # R² score  # Evaluate model using validation fold
        scores.append(score)  # Save the R² score for this fold

    fold_results[name] = scores  # Store all fold scores for the current model

# Plot R² score per fold
plt.figure(figsize=(12, 6))  # Set figure size for better readability

for name, scores in fold_results.items():  # Loop through each model and its fold scores
    plt.plot(range(1, 8), scores, marker='o', label=name)  # Plot R² score from Fold 1 to Fold 7

plt.title('R² Score Per Fold (7-Fold Cross-Validation)')  # Add graph title
plt.xlabel('Fold')  # Label x-axis as fold number
plt.ylabel('R² Score')  # Label y-axis as R² score
plt.xticks(range(1, 8))  # Show fold numbers from 1 to 7
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')  # Place legend outside the graph
plt.grid(alpha=0.3)  # Add light grid lines for easier reading
plt.tight_layout()  # Adjust layout so labels and legend do not overlap
plt.show()  # Display the graph
```

```python
from sklearn.metrics import mean_squared_error, r2_score #naik  # Import metrics to evaluate regression model performance

test_results = {}  # Dictionary to store test results for each model

for name, model in models.items():  # Loop through each model in the models dictionary
    model.fit(X_train, y_train)  # Train the model using the full training dataset
    y_pred = model.predict(X_test)  # Predict target values using the test dataset

    mse = mean_squared_error(y_test, y_pred)  # Calculate Mean Squared Error; lower value means better performance
    r2 = r2_score(y_test, y_pred)  # Calculate R² score; higher value means better model fit

    test_results[name] = {  # Store the evaluation results for the current model
        "MSE": mse,  # Save the model's Mean Squared Error
        "R2": r2  # Save the model's R² score
    }

print(test_results)  # Display all model test results
```

```python
final_results = [] # Create an empty list to store the final evaluation results

for name in models.keys():  # Loop through each model name
    cv_r2 = cv_results[name]["CV_R2_Mean"]  # Get the average R² score from cross-validation
    cv_std = cv_results[name]["CV_R2_STD"]  # Get the standard deviation of R² scores across folds
    test_r2 = test_results[name]["R2"]  # Get the R² score from the test set
    mse = test_results[name]["MSE"]  # Get the Mean Squared Error from the test set
    gap = cv_r2 - test_r2  # Calculate the difference between CV R² and test R²
    rmse = np.sqrt(mse)  # Calculate RMSE from MSE for easier interpretation

    # Simple interpretation rules
    if cv_r2 < 0.2 and test_r2 < 0.2:  # If both CV and test R² are low, the model may not learn the pattern well
        status = "Underfitting"
    elif gap > 0.05:  # If CV R² is much higher than test R², the model may perform worse on unseen data
        status = "Overfitting"
    else:  # If the gap is small and performance is acceptable, the model is considered stable
        status = "Good Fit"

    final_results.append({  # Add the current model's final results into the list
        "Model": name,  # Store model name
        "CV_R2": round(cv_r2, 4),  # Store rounded average cross-validation R²
        "CV_R2_STD": round(cv_std, 4),  # Store rounded variation of CV R² scores
        "Test_R2": round(test_r2, 4),  # Store rounded test R² score
        "MSE": round(mse, 4),  # Store rounded Mean Squared Error
        "RMSE": round(rmse, 4),  # Store rounded Root Mean Squared Error
        "Gap(CV-Test)": round(gap, 4),  # Store rounded gap between CV and test R²
        "Status": status  # Store model condition: Underfitting, Overfitting, or Good Fit
    })

df_results = pd.DataFrame(final_results)  # Convert final results list into a DataFrame table
df_results = df_results.sort_values(by="Test_R2", ascending=False)  # Sort models from highest to lowest test R²

print(df_results)  # Display the final model comparison table
```
