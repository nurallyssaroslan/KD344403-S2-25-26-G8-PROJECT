**Data Cleaning**

```python
# Print a message describing the dataset shape
print("Size of df (rows, columns):")

# Display number of rows and columns in the dataset
display(df.shape)
```

```python
# Check number of duplicate rows in the dataset
print("Number of duplicate rows in df:")

# Count and display duplicate rows
display(df.duplicated().sum())
```

```python
# Remove duplicate rows from the dataset
df = df.drop_duplicates()
```

```python
# Print label describing dataset shape
print("Size of df (rows, columns):")

# Display number of rows and columns in the dataset
display(df.shape)
```

```python
# Drop unnecessary/redundant columns to reduce noise in the dataset
df = df.drop(columns=['salary', 'salary_currency'])
```

```python
# Display the first 5 rows of the dataset
display(df.head())
```

```python
# Print message describing the dataset shape
print("Size of df (rows, columns):")

# Display number of rows and columns in the dataset
display(df.shape)
```

```python
# Convert text columns to lowercase for consistency

df['job_title'] = df['job_title'].str.lower()
df['experience_level'] = df['experience_level'].str.lower()
df['employment_type'] = df['employment_type'].str.lower()
df['employee_residence'] = df['employee_residence'].str.lower()
df['company_location'] = df['company_location'].str.lower()
df['company_size'] = df['company_size'].str.lower()

# Display updated dataset
display(df.head())
```

```python
# checking how remote ratio influences the target
plt.figure(figsize=(10, 6))
sns.boxplot(x='remote_ratio', y='salary_in_usd', data=df)
plt.title('Salary Distribution by Remote Ratio')
plt.xlabel('Remote Ratio (%)')
plt.ylabel('Salary in USD')
plt.grid(axis='y', alpha=0.75)
plt.show()

# the salary distribution between fully remote and on-site are almost similar
# this may indicate that fully remote and on-site does not have a huge effect on the target
# however, hybrid suggest some variation in salary
```

```python
remote_ratio_percentages = df['remote_ratio'].value_counts(normalize=True) * 100
display(remote_ratio_percentages.round(2))

# however, from here we can see that hybrid has too few data to learn reliably
# furthermore, fully remote and on-site show similiar distribution
# decision: low/moderate signal may help the target, further evaluation after model
```

```python
# checking how employement type influences the target
plt.figure(figsize=(10, 6))
sns.boxplot(x='employment_type', y='salary_in_usd', data=df)
plt.title('Salary Distribution by Employment Type')
plt.xlabel('Employment Type')
plt.ylabel('Salary in USD')
plt.grid(axis='y', alpha=0.75)
plt.show()

# pt and fl show similiar salary range
# ct shows a generally higher salary
```

```python
percentages = df['employment_type'].value_counts(normalize=True) * 100

print(percentages.round(2))

# highly imbalance with too few data from ct, pt and fl to learn reliably
# 98.96% is ft, other employement type are almost invisble and may introduce noise to model
# decision: drop employement_type
```

```python
# Drop the 'employment_type' column from the dataset
df = df.drop(columns=['employment_type'])
```

```python
# Display the first 5 rows of the dataset
display(df.head())
```

```python
# Set figure size for the plot
plt.figure(figsize=(10, 6))

# Plot histogram of salary distribution with KDE curve
sns.histplot(df['salary_in_usd'], bins=30, kde=True)

# Add plot title
plt.title('Distribution of Salary in USD')

# Label x-axis
plt.xlabel('Salary in USD')

# Label y-axis
plt.ylabel('Frequency')

# Add grid for better readability
plt.grid(axis='y', alpha=0.75)

# Display the plot
plt.show()
```

```python
# Show full distribution of job titles
print("Distribution of job titles:")

# Ensure all rows are visible in output
pd.set_option('display.max_rows', None)

# Count occurrences of each job title
display(df['job_title'].value_counts())
```

```python
# Set threshold for rare job titles (1% of dataset)
threshold = 0.01

# Calculate proportion of each job title
counts = df['job_title'].value_counts(normalize=True)

# Group rare job titles into 'other'
df['job_title_clean'] = df['job_title'].apply(
    lambda x: x if counts[x] >= threshold else 'other'
)
```

```python
# Remove the 'job_title' column from the dataset
df = df.drop(columns=['job_title'])
```

```python
# Print title for clarity
print("Distribution of job titles:")

# Show all rows when displaying value counts
pd.set_option('display.max_rows', None)

# Count and display frequency of each job title
display(df['job_title_clean'].value_counts())
```

**EDA**

`experience_level` vs `salary_in_usd`

```python
# Create a figure with specific size
plt.figure(figsize=(10, 6))

# Boxplot to show salary distribution by experience level
sns.boxplot(x='experience_level', y='salary_in_usd', data=df, order=['en', 'mi', 'se', 'ex'])

# Add chart title
plt.title('Salary Distribution by Experience Level')

# Label x-axis
plt.xlabel('Experience Level')

# Label y-axis
plt.ylabel('Salary in USD')

# Add horizontal grid lines for better readability
plt.grid(axis='y', alpha=0.75)

# Display the plot
plt.show()
```

Top 10 `employee_residence` vs `salary_in_usd`

```python
# Get top 10 countries by number of employees
top_countries = df['employee_residence'].value_counts().nlargest(10).index

# Filter dataset to only include top 10 countries
df_top = df[df['employee_residence'].isin(top_countries)]

# Set figure size for better readability
plt.figure(figsize=(12,6))

# Boxplot of salary distribution by employee residence
sns.boxplot(x='employee_residence', y='salary_in_usd', data=df_top)

# Add title to the plot
plt.title('Salary Distribution by Employee Residence (Top 10)')

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Calculate mean salary by employee residence (top 10 countries)
mean_salary = df.groupby('employee_residence')['salary_in_usd'].mean() \
                .sort_values(ascending=False).head(10)

# Set figure size for better readability
plt.figure(figsize=(10,6))

# Plot bar chart of top 10 mean salaries
mean_salary.plot(kind='bar')

# Add chart title
plt.title('Mean Salary by Employee Residence (Top 10)')

# Label y-axis
plt.ylabel('Mean Salary (USD)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

Top 10 `company_location` vs `salary_in_usd`


```python
# Get top 10 company locations by frequency
top_companies = df['company_location'].value_counts().nlargest(10).index

# Filter dataset to only include top 10 company locations
df_top = df[df['company_location'].isin(top_companies)]

# Set figure size for better readability
plt.figure(figsize=(12,6))

# Create boxplot of salary by company location
sns.boxplot(x='company_location', y='salary_in_usd', data=df_top)

# Add title to the plot
plt.title('Salary Distribution by Company Location (Top 10)')

# Rotate x-axis labels for clarity
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Calculate mean salary by company location and get top 10 highest
mean_salary = df.groupby('company_location')['salary_in_usd'].mean() \
                .sort_values(ascending=False).head(10)

# Create figure size for the plot
plt.figure(figsize=(10,6))

# Plot bar chart of mean salary by company location
mean_salary.plot(kind='bar')

# Set title of the chart
plt.title('Mean Salary by Company Location (Top 10)')

# Label y-axis
plt.ylabel('Mean Salary (USD)')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

```python
# Calculate average salary by job title
average_salary_by_job_title = (
    df.groupby('job_title_clean')['salary_in_usd']
    .mean()
    .sort_values(ascending=False)
)

# Show all rows in output (no truncation)
pd.set_option('display.max_rows', None)

# Print label
print("Average salary by job title:")

# Display result
display(average_salary_by_job_title)
```
